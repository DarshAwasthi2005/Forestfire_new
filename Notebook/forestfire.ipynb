# LASSO, RIDGE AND ELASTIC NET


```python
df["Classes"].value_counts()
```

```python
df["classes"]=np.where(df["Classes"].str.contains("not fire"),0,1)
```

```python
df["classes"].value_counts()
```

```python
X=df.drop("FWI", axis=1)
```

```python
Y=df["FWI"]
```

```python
X.head()
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=.25, random_state=42)
```

```python
X_train.shape,Y_train.shape
```

```python
X_train.corr
```

```python
plt.figure(figsize=(12,8))
corr=X_train.select_dtypes(include=np.number).corr()
sns.heatmap(corr, annot=True)
```

```python
def correlation(dataset, threshold):
    col_corr = set() 
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold:
                colname = corr_matrix.columns[i] 
                col_corr.add(colname)
                
    return col_corr
```

```python
corr_features = correlation(X_train.select_dtypes(include=np.number), 0.85)
```

```python
X_train.drop(corr_features, axis=1, inplace=True)
X_test.drop(corr_features, axis=1, inplace=True)
X_train.shape,Y_train.shape
```

```python
X_train.columns
X_test.columns
```

```python
X_train=X_train.drop('Classes', axis=1)
X_test=X_test.drop('Classes', axis=1)
```

```python
X_train.head
```

# Feature Scaling and Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
plt.subplots(figsize=(10,5))
plt.subplot(1,2,1)
sns.boxplot(data=X_train_scaled)
plt.title("X_train Scaled")
plt.subplot(1,2,2)
sns.boxplot(data=X_test_scaled)
plt.title("X_test Scaled")
```

# Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
linreg=LinearRegression()
```

```python
linreg.fit(X_train_scaled,Y_train)
Y_pred=linreg.predict(X_test_scaled)
```

```python
mae=mean_squared_error(Y_test,Y_pred)
score=r2_score(Y_test,Y_pred)
print("Mean Absolute Error:", mae)
print("R^2 Score:", score)
```

```python
plt.scatter(Y_test, Y_pred)
```

# Lasso


```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error, r2_score
lasso=Lasso()
lasso.fit(X_train_scaled,Y_train)
Y_pred=lasso.predict(X_test_scaled)
mse=mean_squared_error(Y_test,Y_pred)
score=r2_score(Y_test,Y_pred)
print("Mean Squared Error:", mse)
print("R^2 Score:", score)
plt.scatter(Y_test, Y_pred)
```

# LassoCV


```python
from sklearn.linear_model import LassoCV
lassocv=LassoCV(cv=5)
lassocv.fit(X_train_scaled,Y_train)
```

```python
lassocv.alpha_
```

```python
Y_pred=lassocv.predict(X_test_scaled)
mse=mean_squared_error(Y_test,Y_pred)
score=r2_score(Y_test,Y_pred)
print("Mean Squared Error:", mse)
print("R^2 Score:", score)
```

```python
plt.scatter(Y_test, Y_pred)
```

# Ridge

```python
from sklearn.linear_model import Ridge
ridge=Ridge()
ridge.fit(X_train_scaled,Y_train)
Y_pred=ridge.predict(X_test_scaled)
mse=mean_squared_error(Y_test,Y_pred)
score=r2_score(Y_test,Y_pred)
print("Mean Squared Error:", mse)
print("R^2 Score:", score)
```

```python
plt.scatter(Y_test, Y_pred)
```

```python
from sklearn.linear_model import RidgeCV
ridgecv=RidgeCV(cv=5)
ridgecv.fit(X_train_scaled,Y_train)
Y_pred=ridgecv.predict(X_test_scaled)
mse=mean_squared_error(Y_test,Y_pred)
score=r2_score(Y_test,Y_pred)
print("Mean Squared Error:", mse)
print("R^2 Score:", score)
```

```python
plt.scatter(Y_test, Y_pred)
```
